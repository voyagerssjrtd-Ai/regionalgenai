```python
from bs4 import BeautifulSoup
import requests
from urllib.parse import urljoin
url = 'https://www.geeksforgeeks.org/python/implementing-web-scraping-python-beautiful-soup/'
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")

# Step 2: Extract article text
# This depends on the site's structure. Common tags: <article>, <div class="content">, etc.
article = soup.find("article") or soup.find("div", class_="content")
if article:
    text = article.get_text(separator="\n", strip=True)
    print("üìù Article Text:\n", text)
else:
    print("No article content found.")
```

```python
# Step 3: Extract image URLs
images = []
for img in soup.find_all("img"):
    src = img.get("src")
    if src:
        full_url = urljoin(url, src)  # Handle relative URLs
        images.append(full_url)

print("\nüñºÔ∏è Image URLs:")
for img_url in images:
    print(img_url)
```
