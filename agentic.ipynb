```python
from langchain_openai import ChatOpenAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
import os
import httpx
from langchain_community.document_loaders import PyPDFLoader
from langchain_experimental.text_splitter import SemanticChunker
from langchain_openai import OpenAIEmbeddings

loader = PyPDFLoader(r"c:\Users\Genaiblrpiousr13\Downloads\test\checking\files\photos.pdf")
docs = loader.load()
full_text = "\n".join([doc.page_content for doc in docs])

prompt = PromptTemplate.from_template("""
Split the following document into semantically meaningful sections. 
Each section should be self-contained and preserve context. 
Return the result as a list of chunks.

Document:
{document}
""")

chunking_chain = LLMChain(llm=llm, prompt=prompt)
response = chunking_chain.run({"document": full_text})

# # chunks
```

```python
chunks = response.split("\n\n")  # or use a smarter parser
```

```python
from langchain.schema import Document
from langchain_community.vectorstores import Chroma
chunks = [chunk.strip() for chunk in response.split("\n\n") if chunk.strip()]
chunked_docs = [Document(page_content=chunk) for chunk in chunks]

# âœ… Step 4: Store in ChromaDB
embedding_model = OpenAIEmbeddings()
vector_store = Chroma.from_documents(
    chunked_docs,
    embedding_model,
    persist_directory="chroma_db"
)
vector_store.persist()
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
import os

# Load PDF
loader = PyPDFLoader(r"c:\Users\Genaiblrpiousr13\Downloads\test\checking\files\constitution_of_india.pdf")
docs = loader.load()

# Set your OpenAI API key
os.environ["OPENAI_API_KEY"] = "sk-R9XkOr0HBHivohV4WyK6iQ"

# Split the documents into chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)
split_docs = text_splitter.split_documents(docs)

# Embed the chunks
embedding_model = OpenAIEmbeddings()
embeddings = embedding_model.embed_documents([doc.page_content for doc in split_docs])

print(f"âœ… Split into {len(split_docs)} chunks and embedded successfully.")
```
