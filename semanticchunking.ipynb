```python
from langchain_community.document_loaders import PyMuPDFLoader, DirectoryLoader
dir = DirectoryLoader(
    "c:/Users/Lenovo/OneDrive/Desktop/GenAI/VoyagersRegionalGenAI/savedFiles",
    glob = "**/*.pdf",
    loader_cls = PyMuPDFLoader,
    show_progress = False
)
pdf = dir.load()
all_texts = [doc.page_content for doc in pdf]
all_texts
```

```python
import os
from langchain_experimental.text_splitter import SemanticChunker
from langchain_openai import OpenAIEmbeddings

# üîê Set your OpenAI API key
os.environ["OPENAI_API_KEY"] = "your-api-key-here"  # Replace with your actual key

# üß† Initialize the semantic chunker
text_splitter = SemanticChunker(
    OpenAIEmbeddings(),
    breakpoint_threshold_type="percentile"  # Options: "standard_deviation", "interquartile"
)

# üìÑ Create documents from your list of texts
documents = text_splitter.create_documents(all_texts)
```
