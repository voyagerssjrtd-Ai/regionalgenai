```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#creating the random data
def generate_data(size = 1000):
    np.random.seed(42)
    time=pd.date_range(start = '2025-01-01', periods=size,freq ='h')
    data = np.random.normal(loc=50, scale=5,size=size)
    data[100:120] +=20
    data[500:520] -=30
    df = pd.DataFrame({'Time':time, "Value":data})
    return df
```

```python
df = generate_data()
df.head()
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
# Preprocessing the data
df['Time'] = pd.to_datetime(df['Time'])
df.set_index('Time',inplace = True)
```

```python
df = df.ffill()
```

```python
from sklearn.preprocessing import StandardScaler
scaler  = StandardScaler()
df['Scaled_Value'] = scaler.fit_transform(df[['Value']])
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(contamination=0.05)
df['Anomaly'] = model.fit_predict(df[['Scaled_Value']])

df['Anomaly'] = df['Anomaly'].apply(lambda x : 'Anomaly' if x == -1 else 'Normal')

print(df.tail())
```

```python
plt.figure(figsize=(14,7))
sns.lineplot(data=df, x = 'Time', y='Value', label = 'Normal Data', color = 'blue')
sns.scatterplot(data=df[df['Anomaly'] == 'Anomaly'], x = 'Time', y='Value', label = "Anomalies", color = 'red')
plt.title("Real-Time Anomaly Detection")
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
import time

for i in range(3):
    new_data = generate_data(size = 100)
    df = pd.concat([df,new_data],ignore_index=True)
    df['Scaled_Value'] = scaler.fit_transform(df[['Value']])
    df['Anomaly'] = model.fit_predict(df[['Scaled_Value']])
    df['Anomaly'] = df['Anomaly'].apply(lambda x : 'Anomaly' if x == -1 else 'Normal')
    plt.figure(figsize=(12,7))
    sns.lineplot(data=df, x = 'Time', y='Value', label = 'Normal Data', color = 'blue')
    sns.scatterplot(data=df[df['Anomaly'] == 'Anomaly'], x = 'Time', y='Value', label = "Anomalies", color = 'red')
    plt.title("Real-Time Anomaly Detection")
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.legend()
    plt.show()

    time.sleep(2)
    
```
